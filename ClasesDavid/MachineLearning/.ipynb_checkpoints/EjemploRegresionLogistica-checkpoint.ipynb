```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Crear datos de ejemplo
np.random.seed(42)
num_samples = 1500
exam_scores = np.random.rand(num_samples) * 100
study_hours = np.random.rand(num_samples) * 10
admission_status = (0.3 * exam_scores + 0.5 * study_hours + np.random.rand(num_samples) * 10) > 15
```

```python
exam_scores
```

```python
admission_status
```

```python
study_hours
```

```python
# Crear un DataFrame de Pandas
data = pd.DataFrame({
    'exam_scores': exam_scores,
    'study_hours': study_hours,
    'admission_status': admission_status
})
```

```python
data
```

```python

# Dividir los datos en conjuntos de entrenamiento y prueba
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)
```

```python
train_data
```

```python
test_data
```

```python
# Definir las variables predictoras (features) (X) y la variable objetivo(target) (y)
X_train = train_data[['exam_scores', 'study_hours']]
y_train = train_data['admission_status']

X_test = test_data[['exam_scores', 'study_hours']]
y_test = test_data['admission_status']
```

```python
X_test
```

```python
type(X_train)
```

```python
# Crear el modelo de regresión logística
model = LogisticRegression()

# Entrenar el modelo
model.fit(X_train, y_train)
```

```python
# Realizar predicciones en el conjunto de prueba
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
# Calcular la precisión del modelo
accuracy = accuracy_score(y_test, y_pred)
print("Precisión del modelo:", accuracy)
```

```python

# Visualizar los resultados
plt.scatter(data['exam_scores'], data['study_hours'], c=data['admission_status'], cmap='coolwarm', marker='o')
plt.xlabel("Puntuación de Exámenes")
plt.ylabel("Horas de Estudio")
plt.title("Admisión a la Universidad")
plt.show()
```
