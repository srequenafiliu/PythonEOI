# Regresión Lineal:


En este ejercicio se explican los fundamentos básicos de la regresión lineal

### 1. Generación del conjunto de datos

```python
import numpy as np

X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

print("La longitud del conjunto de datos es:", len(X))
```

```python
type(X)
```

```python
X
```

### 2. Visualización del conjunto de datos

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(X, y, "b.")
plt.show()
```

```python
plt.plot(X, y, "b.")
plt.xlabel("Variable Independiente")
plt.ylabel("Variable Dependiente")
plt.show()
```

### 3. Modificación del conjunto de datos

```python
X.flatten()
```

```python
import pandas as pd
```

```python
data = {'vi': X.flatten(), 'vd': y.flatten()}
df = pd.DataFrame(data)
df.head(10)
```

### 4. Construcción del modelo

```python
# pip install scikit-learn
from sklearn.linear_model import LinearRegression
```

```python
df['vi'].values.reshape(-1, 1)
```

```python
# Construcción del modelo y ajuste de la función hipótesis
lin_reg = LinearRegression()
lin_reg.fit(df['vi'].values.reshape(-1, 1), df['vd'].values)
```

```python
# Parámetro theta 0
lin_reg.intercept_
```

```python
# Parámetro theta 1
lin_reg.coef_
```

```python
# Predicción para el valor mínimo y máximo del conjunto de datos de entrenamiento
X_min_max = np.array([[df["vi"].min()], [df["vi"].max()]])
y_train_pred = lin_reg.predict(X_min_max)
```

```python
y_train_pred
```

```python
X_min_max
```

```python
# Representación gráfica de la función hipótesis generada
plt.plot(X_min_max, y_train_pred, "g-")
plt.plot(df['vi'], df['vd'], "b.")
plt.xlabel("Variable Independiente")
plt.ylabel("Variable Dependiente.")
plt.show()
```

### 5. Predicción de nuevos ejemplos

```python
x_new = np.array([[0.75]]) 

# Predicción del coste que tendría el incidente
vd = lin_reg.predict(x_new) 

print("El valor de la variable Dependiente será", int(vd[0]), "€")
```
