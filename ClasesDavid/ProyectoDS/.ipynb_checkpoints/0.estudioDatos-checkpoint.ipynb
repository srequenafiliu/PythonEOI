```python
# Todas las importaciones
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("HousePrices_HalfMil.csv")
df.head()
```

```python
df.info()
```

```python
var_num = df[['Area', 'Garage', 'FirePlace', 'Baths']]
var_cat = df.drop(['Area', 'Garage', 'FirePlace', 'Baths', 'Prices'], axis=1)
```

```python
def regresionLineal(X, Y, tagX, tagY):
    model = LinearRegression()
    model.fit(X.values.reshape(-1, 1), Y)
    pendiente = model.coef_[0] # .coef_ obtiene las pendientes del modelo, una para cada columna
    ordenada = model.intercept_ # Obtiene la ordenada en el origen
    # Crear puntos para la línea de regresión
    x_linea = np.linspace(min(X), max(X), 100)
    y_linea = pendiente * x_linea + ordenada
    plt.plot(x_linea, y_linea)
    plt.xlabel(tagX)
    plt.ylabel(tagY)
    plt.title(f'Relación {tagX}-{tagY}')
    plt.show()
```

```python
for col in var_num.columns:
    regresionLineal(df[col], df['Prices'], col, 'Prices')
```

```python
def boxplot(data, tagX, tagY):
    sns.boxplot(x=tagX, y=tagY, data=data)
    plt.xlabel(tagX)
    plt.ylabel(tagY)
    plt.title(f'Relación {tagX}-{tagY}')
    plt.show()
```

```python
for col in var_cat.columns:
    boxplot(df, col, 'Prices')
```

```python
def pendiente(col):
    model = LinearRegression()
    model.fit(df[col].values.reshape(-1, 1), df['Prices'])
    return model.coef_[0]
```

```python
# Como algunos diagramas de cajas no son claros, se usan las pendientes para ver la relación
datos_no_claros = ['Black Marble', 'Solar', 'Electric', 'Swiming Pool', 'Garden']
for col in datos_no_claros:
    pendiente_dato = pendiente(col)
    print(f"Pendiente de la característica {col}: {pendiente_dato:.2f}")
```
