```python
# Todas las importaciones
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("HousePrices_HalfMil.csv")
df.head()
```

```python
df.info()
```

```python
y = df['Prices']
y.head()
```

```python
X = df.drop('Prices', axis=1)
X.head()
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
coefficients = model.coef_ # Obtiene las pendientes del modelo, una para cada columna
intercept = model.intercept_ # Obtiene la ordenada en el origen
coefficients, intercept
```

```python
# Como las pendientes de Swiming Pool y Garden son despreciables, se vuelven a calcular las pendientes y la ordenada en el origen
X = df.drop(['Swiming Pool', 'Garden', 'Prices'], axis=1)
X.head()
```

```python
model = LinearRegression()
model.fit(X, y)
coefficients = model.coef_ # Obtiene las pendientes del modelo, una para cada columna
intercept = model.intercept_ # Obtiene la ordenada en el origen
coefficients, intercept
```

```python
def pendiente(valores_x):
    pendiente_general = 0
    for coef, x in zip(coefficients, valores_x):
        pendiente_general += coef*x
    return pendiente_general

def precio_teorico(fila):
    return int(pendiente(fila) + intercept)
```
