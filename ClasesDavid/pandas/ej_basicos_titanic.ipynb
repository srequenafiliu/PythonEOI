```python
import pandas as pd
df = pd.read_csv("titanic3.csv")
# df = pd.read_csv("titanic3.csv", dtype={"age":float})
df
```

```python
df['age'] = df["age"].replace({",": "."}, regex=True).astype(float)
df.info()
```

Seleccionar todos los pasajeros que eran mujeres

```python
df.loc[df['sex'] == 'female']
```

Seleccionar todos los pasajeros que tenían entre 18 y 30 años.

```python
df.loc[(df['age'] >= 18) & (df['age'] <= 30)]
```

Seleccionar todos los pasajeros que eran de primera clase.

```python
first_class = df.loc[(df['pclass'] == 1)]
first_class
```

Seleccionar todos los pasajeros que sobrevivieron al accidente.

```python
survived_passengers = df.loc[df['survived'] == 1]
survived_passengers
```

Seleccionar todos los pasajeros que murieron en el accidente.

```python
no_survived_passengers = df.loc[df['survived'] == 0]
no_survived_passengers
```

¿Qué porcentaje de pasajeros sobrevivieron al accidente?

```python
len(survived_passengers) * 100 / len(df)
```

¿Cuál era la edad promedio de los pasajeros que sobrevivieron?

```python
survived_passengers['age'].mean()
```

¿Cuál era la edad promedio de los pasajeros que no sobrevivieron?

```python
no_survived_passengers['age'].mean()
```

¿Qué porcentaje de pasajeros de primera clase sobrevivieron?

```python
survived_1class = df.loc[(df['survived'] == 1) & (df['pclass'] == 1)]
len(survived_1class) * 100 / len(df)
```

¿Cuál era la probabilidad de supervivencia para los pasajeros de primera clase, segunda clase y tercera clase?

```python
survived_first_class = first_class.loc[df['survived'] == 1]
second_class = df.loc[df['pclass'] == 2]
survived_second_class = second_class.loc[df['survived'] == 1]
third_class = df.loc[df['pclass'] == 3]
survived_third_class = third_class.loc[df['survived'] == 1]
print(f"Probabilidad para los pasajeros de primera clase: {len(survived_first_class) * 100 / len(first_class):.2f}%")
print(f"Probabilidad para los pasajeros de segunda clase: {len(survived_second_class) * 100 / len(second_class):.2f}%")
print(f"Probabilidad para los pasajeros de tercera clase: {len(survived_third_class) * 100 / len(third_class):.2f}%")
```

¿Hubo diferencias en las tasas de supervivencia entre hombres y mujeres?

```python
df.groupby('sex')['survived'].mean().std() != 0
```

¿Hubo diferencias en las tasas de supervivencia entre pasajeros de diferentes edades?

```python
survival_by_age_group = pd.cut(df['age'], bins=[0, 12, 18, 30, 60, 100], labels=['child', 'teen', 'young_adult', 'adult', 'elderly'])
df.groupby(survival_by_age_group)['survived'].mean().std() != 0
```

¿Hubo diferencias en las tasas de supervivencia entre pasajeros de diferentes clases?

```python
df.groupby('pclass')['survived'].mean().std() != 0
```
