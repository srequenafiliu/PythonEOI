Antes de avanzar necesitamos aprender una cosa de Pandas.

```python
import pandas as pd

data = {
    'alimento': ['Manzana', 'Chocolate', 'Brócoli', 'Galletas', 'Pescado'],
    'salud': ['Saludable', 'No saludable', 'Saludable', 'No saludable', 'Saludable']
}

df = pd.DataFrame(data)

# Crear un mapping para la transformación
mapping_salud = {'Saludable': 1, 'No saludable': 0}

# Aplicar el mapeo a la columna 'salud'
df['salud_numerica'] = df['salud'].map(mapping_salud)

print(df)
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats

np.random.seed(42)
n = 50
data = {'nombre': ['Juan', 'María', 'Carlos', 'Laura', 'Pedro'] * 10,
        'edad': np.random.randint(18, 65, n),
        'peso': np.random.randint(50, 100, n),
        'nivel_educacion': np.random.choice(['Primaria', 'Secundaria', 'Universidad'], n),
        'nivel_ingresos': np.random.choice(['Bajo', 'Medio', 'Alto'], n),
        'satisfaccion': np.random.choice(['Baja', 'Media', 'Alta'], n)}

df = pd.DataFrame(data)
```

```python
df.head(50)
```

```python
cross_tab_educ_ingresos = pd.crosstab(df['nivel_educacion'], df['nivel_ingresos'])
cross_tab_educ_ingresos
```

```python
# Chi-cuadrado (Nivel de Educación y Nivel de Ingresos)

chi2, p_value, _, _ = stats.chi2_contingency(cross_tab_educ_ingresos)
print("Valor de chi-cuadrado (Nivel de Educación y Nivel de Ingresos):", chi2)
print("Valor p:", p_value)
```
