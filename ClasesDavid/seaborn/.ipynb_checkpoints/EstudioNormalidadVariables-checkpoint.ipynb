### Estudio de la normalidad de una variable.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
np.random.seed(890)

x = np.random.normal(loc=0, scale=1, size=100)

plt.hist(x, bins=10)
plt.title("Histograma de una variable normal")
plt.xlabel("Valor de la variable")
plt.ylabel("Frecuencia")
plt.show()
```

### Gráficos QQ

Los gráficos Q-Q (quantile-quantile) son otra forma de comparar la distribución de una variable con una distribución teórica, como la normal. Consisten en representar los cuantiles de la variable observada frente a los cuantiles de la distribución teórica. Si la variable sigue la misma distribución que la teórica, los puntos se alinearán aproximadamente sobre una recta. Por ejemplo, si usamos la misma muestra que antes, podemos obtener un gráfico Q-Q como este:

```python
stats.probplot(x, dist="norm", plot=plt)
plt.title("Gráfico Q-Q de una variable normal")
plt.xlabel("Cuantiles teóricos")
plt.ylabel("Cuantiles observados")
plt.show()
```

### Test de Shapiro-Wilk

La prueba de Shapiro-Wilk es un test estadístico que contrasta la hipótesis nula de que una variable sigue una distribución normal contra la hipótesis alternativa de que no lo hace. El test devuelve un valor p que indica la probabilidad de obtener los datos observados si la hipótesis nula fuera cierta. Si el valor p es menor que un nivel de significancia preestablecido (por ejemplo, 0.05), se rechaza la hipótesis nula y se concluye que la variable no sigue una distribución normal. Por ejemplo, si aplicamos el test a la misma muestra que antes, obtenemos el siguiente resultado

```python
# Aplicamos el test de Shapiro-Wilk
stat, p = stats.shapiro(x)
print(f"Estadístico: {stat:.3f}")
print(f"Valor p: {p:.3f}")
```

Como el valor p es mayor que 0.05, no podemos rechazar la hipótesis nula y asumimos que la variable sigue una distribución normal.
