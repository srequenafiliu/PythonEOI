```python
quejas_por_estado = df.groupby('ESTADO_PROCESAL')['ID_EXP'].count()
```

```python
import matplotlib.pyplot as plt
quejas_por_estado.plot(kind='pie'  , labels=None, autopct='%1.1f%%',figsize=(10,10))

plt.legend(labels=quejas_por_estado.index, loc='upper right')

plt.title('Porcentaje de quejas por estado procesal')

plt.show()
```

```python
import datetime
tiempo_medio = df['FECHA DE CIERRE'].mean(axis=0)
tiempo_medio = tiempo_medio.date()
tiempo_medio = datetime.date.today() - tiempo_medio
tiempo_medio = tiempo_medio.days
print(f'El tiempo medio de cierre de las quejas es de: {tiempo_medio} días')
```

```python
import seaborn as sns
sum_casos = df["ESTADO_PROCESAL"].value_counts()

plt.pie(sum_casos, labels=sum_casos.index, autopct="%0.1f %%", colors=sns.color_palette("pastel"))

plt.tight_layout()
```

```python
df['PERDIDA'] = df['MONTO RECLAMADO'] - df['MONTO RECUPERADO'] 
df_perdida = df.groupby('NOMBRE_COMERCIAL')['PERDIDA'].sum().reset_index()
```

```python
df_perdida = df_perdida.sort_values(by='PERDIDA', ascending=False).head(10)
```

```python
ax = df_perdida.plot.bar(x='NOMBRE_COMERCIAL', y='PERDIDA', rot=90, figsize=(10, 6)) 
ax.set_xlabel('Empresa')
ax.set_ylabel('Pérdida') 
ax.set_title('Pérdida por empresa')
plt.show()
```
