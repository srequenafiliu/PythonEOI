## Estudia la normalidad de las variables del dataset "planets" que viene en seaborn

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns

planets = sns.load_dataset('planets')
planets.info()
```

```python
def calcular_normalidad(x):
    stats.probplot(x, dist="norm", plot=plt)
    plt.title("Gráfico Q-Q")
    plt.xlabel("Cuantiles teóricos")
    plt.ylabel("Cuantiles observados")
    plt.show()
    stat, p = stats.shapiro(x)
    print(f"Estadístico: {stat:.3f}")
    print(f"Valor p: {p:.3f}")
```

```python
# Estudio de normalidad para el periodo orbital
planets = planets.dropna(subset=['orbital_period']) # Quitar valores nulos
calcular_normalidad(planets['orbital_period'])
```

```python
# Estudio de normalidad para la masa
planets = planets.dropna(subset=['mass']) # Quitar valores nulos
calcular_normalidad(planets['mass'])
```

```python
# Estudio de normalidad para la distancia
planets = planets.dropna(subset=['distance']) # Quitar valores nulos
calcular_normalidad(planets['distance'])
```
