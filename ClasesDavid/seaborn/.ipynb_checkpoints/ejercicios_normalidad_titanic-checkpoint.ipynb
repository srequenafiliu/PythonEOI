## Estudia la normalidad de las variables del dataset del titanic.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns

titanic = sns.load_dataset('titanic')
titanic.info()
```

```python
def eliminar_outliers(x):
    Q1 = x.quantile(0.25)
    Q3 = x.quantile(0.75)
    IQR = Q3 - Q1
    lim_inf = Q1 - 1.5 * IQR
    lim_sup = Q3 + 1.5 * IQR
    media = x.mean()
    x = x.mask(x > lim_sup, media)
```

```python
def calcular_normalidad(x):
    stats.probplot(x, dist="norm", plot=plt)
    plt.title("Gráfico Q-Q")
    plt.xlabel("Cuantiles teóricos")
    plt.ylabel("Cuantiles observados")
    plt.show()
    stat, p = stats.shapiro(x)
    print(f"Estadístico: {stat:.3f}")
    print(f"Valor p: {p:.3f}")
```

```python
# Estudio de normalidad para la edad
titanic = titanic.dropna(subset=['age']) # Quitar valores nulos
calcular_normalidad(titanic['age'])
```

```python
# Estudio de normalidad para el precio del billete
titanic = titanic.dropna(subset=['fare']) # Quitar valores nulos
eliminar_outliers(titanic['fare'])
calcular_normalidad(titanic['fare'])
```
