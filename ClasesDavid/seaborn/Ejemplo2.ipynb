```python
import pandas as pd
# Crear un diccionario con los datos
datos = {
    "nombre": ["Ana", "Beto", "Carlos", "Diana", "Elena"],
    "edad": [25, 32, 28, 24, 29],
    "sexo": ["F", "M", "M", "F", "F"],
    "altura": [1.65, 1.78, 1.72, 1.60, 1.68],
    "peso": [60, 80, 75, 55, 65],
    "fuma": [False, True, False, False, True]
}
# Crear un DataFrame con el diccionario
df = pd.DataFrame(datos)
# Mostrar el DataFrame
df
```

```python
import seaborn as sns
```

```python
# Crear un histograma de la columna edad
sns.histplot(data=df, x="edad")
```

```python
# Ahora vamos a ver la relación entre dos variables numéricas, como la altura y el peso.
# Crear un diagrama de dispersión de las columnas altura y peso
sns.scatterplot(data=df, x="altura", y="peso")
```

```python
# Relación entre variable numérica y otra categórica
# Crear un diagrama de cajas de las columnas peso y sexo
sns.boxplot(data=df, x="sexo", y="peso")
```

```python
# Relación entre variable numérica y otra categórica
# Crear un diagrama de cajas de las columnas peso y sexo
sns.violinplot(data=df, x="fuma", y="altura")
```

```python
df = df.append({"nombre": "Fernando", "edad": 35, "sexo": "M", "altura": 1.65, "peso": 90, "fuma": True}, ignore_index=True)
df = df.append({"nombre": "Gonzalo", "edad": 40, "sexo": "M", "altura": 1.70, "peso": 95, "fuma": False}, ignore_index=True)
# Mostrar el DataFrame actualizado
df
```

```python
# Crear un diccionario con los datos
datos_ext = {
    "nombre": ["Fernando", "Gonzalo"],
    "edad": [35, 40],
    "sexo": ["M", "M"],
    "altura": [1.65, 1.70],
    "peso": [90, 95],
    "fuma": [True,False]
}
# Crear un DataFrame con el diccionario
df_ext = pd.DataFrame(datos_ext)
df_ext
```

```python
# Crear un diccionario con los datos
datos = {
    "nombre": ["Ana", "Beto", "Carlos", "Diana", "Elena"],
    "edad": [25, 32, 28, 24, 29],
    "sexo": ["F", "M", "M", "F", "F"],
    "altura": [1.65, 1.78, 1.72, 1.60, 1.68],
    "peso": [60, 80, 75, 55, 65],
    "fuma": [False, True, False, False, True]
}
# Crear un DataFrame con el diccionario
df = pd.DataFrame(datos)
# Mostrar el DataFrame
df
```

```python
df_completo = pd.concat([df,df_ext],ignore_index=True)
df_completo
```

```python
import matplotlib.pyplot as plt
# Crear una nueva columna con el IMC de cada persona
df_completo["IMC"] = df_completo.apply(lambda row: row["peso"] / (row["altura"] ** 2), axis=1)

ax = sns.barplot(data=df_completo, x="nombre", y="IMC")
ax.axhline(30, color="red", linestyle="--")
ax.set_title("Índice de masa corporal por persona")
ax.set_xlabel("Nombre")
ax.set_ylabel("IMC")
```

```python
# Contar el número de personas que fuman y que no fuman
fumadores = df["fuma"].value_counts()
ax = sns.barplot(x=fumadores, y=["Fuma", "No fuma"], orient="h")
ax.set_title("Número de personas que fuman y que no fuman")
ax.set_xlabel("Número")
ax.set_ylabel("Fuma")
```

```python
# Contar el número de personas que fuman y que no fuman
fumadores = df["fuma"].value_counts()
# Crear un gráfico de pastel con los valores obtenidos
plt.pie(fumadores, autopct="%1.1f%%", labels=["Fuma", "No fuma"])
# Añadir un título al gráfico
plt.title("Porcentaje de personas que fuman y que no fuman")
```
