```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Cargar el conjunto de datos del Titanic
df = pd.read_excel('titanic3.xls')
df = df.dropna(subset=['age'])
```

```python
sns.set_palette("pastel") # Establecemos una paleta de colores pastel para hacerlo interesante
sns.countplot(x='survived', data=df)
plt.xlabel('Sobrevivieron')
plt.ylabel('Cantidad de Pasajeros')
plt.title('Distribuci√≥n de Sobrevivientes en el Titanic')
plt.show()
```

```python
sns.boxplot(x='pclass', y='age', data=df, palette="viridis")
plt.xlabel('Clase de Billete')
plt.ylabel('Edad')
plt.title('Distribuci√≥n de Edades por Clase de Billete')
plt.show()
```

```python
sns.violinplot(x='pclass', y='age', data=df, palette="Set2", split=True)
plt.xlabel('Clase de Billete')
plt.ylabel('Edades')
plt.title('Distribuci√≥n de Edades por Clase de Billete y G√©nero')
plt.legend(title='G√©nero', loc='upper right')
plt.show()
```

```python

# Filtramos solo los pasajeros que son ni√±os (menores de 18 a√±os)
titanic_ni√±os = df[df['age'] < 15]

# Creamos el gr√°fico de viol√≠n con seaborn
sns.violinplot(x='pclass', y='fare', hue='survived', data=df, split=True)

# A√±adimos los t√≠tulos y etiquetas
plt.title('Distribuci√≥n del precio del billete seg√∫n la clase y la supervivencia de los ni√±os del Titanic')
plt.xlabel('Clase')
plt.ylabel('Precio del billete')
plt.legend(title='Supervivencia', labels=['No', 'S√≠'])

# Mostramos el gr√°fico
plt.show()
```

```python
# Filtramos solo los pasajeros que son ni√±os (menores de 18 a√±os) y que sobrevivieron
df = df[(df['age'] < 18) & (df['survived'] == 1)]

# Creamos una nueva columna con los grupos de edad
df['Age_group'] = pd.cut(df['age'], bins=[0, 5, 10, 15, 18], labels=['0-5', '5-10', '10-15', '15-18'])

# Creamos el gr√°fico de viol√≠n con seaborn
sns.violinplot(x='Age_group', y='fare', data=df,hue="sex")

# A√±adimos los t√≠tulos y etiquetas
plt.title('Distribuci√≥n del precio del billete seg√∫n el grupo de edad de los ni√±os que sobrevivieron en el Titanic')
plt.xlabel('Grupo de edad')
plt.ylabel('Precio del billete')

# Mostramos el gr√°fico
plt.show()
```

```python
# Creamos el gr√°fico de tipo 'reg' con seaborn
sns.jointplot(data=df, x='age', y='fare', kind='reg')

# Mostramos el gr√°fico
plt.show()
```
