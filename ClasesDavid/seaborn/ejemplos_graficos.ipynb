```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Cargar el conjunto de datos del Titanic
df = pd.read_excel('titanic3.xls')
df = df.dropna(subset=['age'])
```

```python
sns.set_palette("pastel") # Establecemos una paleta de colores pastel para hacerlo interesante
sns.countplot(x='survived', data=df)
plt.xlabel('Sobrevivieron')
plt.ylabel('Cantidad de Pasajeros')
plt.title('Distribución de Sobrevivientes en el Titanic')
plt.show()
```

```python
sns.boxplot(x='pclass', y='age', data=df, palette="viridis")
plt.xlabel('Clase de Billete')
plt.ylabel('Edad')
plt.title('Distribución de Edades por Clase de Billete')
plt.show()
```

```python
sns.violinplot(x='pclass', y='age', data=df, palette="Set2", split=True)
plt.xlabel('Clase de Billete')
plt.ylabel('Edades')
plt.title('Distribución de Edades por Clase de Billete y Género')
plt.legend(title='Género', loc='upper right')
plt.show()
```

```python

# Filtramos solo los pasajeros que son niños (menores de 18 años)
titanic_niños = df[df['age'] < 15]

# Creamos el gráfico de violín con seaborn
sns.violinplot(x='pclass', y='fare', hue='survived', data=df, split=True)

# Añadimos los títulos y etiquetas
plt.title('Distribución del precio del billete según la clase y la supervivencia de los niños del Titanic')
plt.xlabel('Clase')
plt.ylabel('Precio del billete')
plt.legend(title='Supervivencia', labels=['No', 'Sí'])

# Mostramos el gráfico
plt.show()
```

```python
# Filtramos solo los pasajeros que son niños (menores de 18 años) y que sobrevivieron
df = df[(df['age'] < 18) & (df['survived'] == 1)]

# Creamos una nueva columna con los grupos de edad
df['Age_group'] = pd.cut(df['age'], bins=[0, 5, 10, 15, 18], labels=['0-5', '5-10', '10-15', '15-18'])

# Creamos el gráfico de violín con seaborn
sns.violinplot(x='Age_group', y='fare', data=df,hue="sex")

# Añadimos los títulos y etiquetas
plt.title('Distribución del precio del billete según el grupo de edad de los niños que sobrevivieron en el Titanic')
plt.xlabel('Grupo de edad')
plt.ylabel('Precio del billete')

# Mostramos el gráfico
plt.show()
```

```python
# Creamos el gráfico de tipo 'reg' con seaborn
sns.jointplot(data=df, x='age', y='fare', kind='reg')

# Mostramos el gráfico
plt.show()
```
