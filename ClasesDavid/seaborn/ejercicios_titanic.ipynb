```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic = sns.load_dataset("titanic")
Q1 = titanic['fare'].quantile(0.25)
Q3 = titanic['fare'].quantile(0.75)
IQR = Q3 - Q1
lim_inf = Q1 - 1.5 * IQR
lim_sup = Q3 + 1.5 * IQR
media_fare = titanic['fare'].mean()
titanic['fare'] = titanic['fare'].mask(titanic['fare'] > lim_sup, media_fare)
titanic.info()
```

Crea un histograma con la distribuci贸n de las edades de los pasajeros

```python
sns.set_palette("pastel")
sns.histplot(x='age', data=titanic, kde=True)
plt.xlabel('Edad')
plt.ylabel('Cantidad de pasajeros')
plt.title('Distribuci贸n de edades por edad')
plt.show()
```

Crea un diagrama de cajas con la distribuci贸n del precio del billete por clase y por sexo
